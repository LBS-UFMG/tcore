```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Read the original data
data = pd.read_csv('/home/usuario-leticia/Desktop/Samuel/leticiaag/tcore/data/processed-data/raw-data1-1km.csv')

# Define trial numbers to remove
trials_to_remove = [68, 32, 1, 71, 16]

# Remove the specified trials
data_filtered = data[~data['trial_number'].isin(trials_to_remove)]

# Print the shape before and after filtering
print(f"Original dataset shape: {data.shape}")
print(f"Filtered dataset shape: {data_filtered.shape}")

# Verify removed trials
print("\nRemoved trials verification:")
for trial in trials_to_remove:
    count = len(data[data['trial_number'] == trial])
    print(f"Trial {trial}: {count} rows removed")

# Update the data variable with filtered data
data = data_filtered
```
